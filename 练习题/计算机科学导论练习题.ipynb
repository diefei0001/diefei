# 计算机科学导论-*Task2：冯诺伊曼模型各部分的关系*
>简述冯诺依曼模型中各部分的作用
>画出冯诺伊曼模型的示意图，揭示各个系统之间的关系
*答案*
[ALU：进行逻辑运算和算术运算;CU：控制其他系统之间的运行与协作;存储器：存储计算机所需的程序和数据;输入/输出：将外部信息转换为二进制数据供计算机处理，并将处理结果转换成人可读的形式]

# 计算机科学导论-*Task3：冯诺依曼模型与图灵机思想*
>冯诺依曼模型是如何体现出图灵机思想的？
*答案*
[ALU：计算能力完整；CU（指令集）：分支选择，无限循环；存储器：动态存储]
